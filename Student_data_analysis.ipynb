```python
names = [s['name'] for s in students]
math_scores = [s['math'] for s in students]
science_scores = [s['science'] for s in students]
english_scores = [s['english'] for s in students]
average_scores = [(s['math'] + s['science'] + s['english'])/3 for s in students]
x = np.arange(len(names))
width = 0.25
```

```python
# 1.Simple Bar Chart
plt.figure(figsize=(10,6))
plt.bar(names, math_scores, color='skyblue')
plt.title("Math Scores of Students")
plt.xlabel("Students")
plt.ylabel("Math Scores")
plt.show()
```

```python
plt.barh(names, math_scores, color='purple')
plt.title("Horizontal Bar Chart - Math Scores")
plt.xlabel("Scores")
plt.ylabel("Students")
plt.show()
```

```python
# 2. Multiple Bar Chart
plt.figure(figsize=(12,6))
plt.bar(x - width, math_scores, width, label='Math')
plt.bar(x, science_scores, width, label='Science')
plt.bar(x + width, english_scores, width, label='English')
plt.xticks(x, names)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Comparison of Scores by Subject")
plt.legend()
plt.show()
```

```python
# 3. Line Plot
plt.figure(figsize=(10,6))
plt.plot(names, math_scores, marker='o', label='Math')
plt.plot(names, science_scores, marker='s', label='Science')
plt.plot(names, english_scores, marker='^', label='English')
plt.title("Scores Trend")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

```python
# 4. Pie Chart
plt.figure(figsize=(8,8))
plt.pie(average_scores, labels=names, autopct='%1.1f%%', startangle=90)
plt.title("Average Scores Distribution Among Students")
plt.show()
```

```python
# 5. Histogram
plt.figure(figsize=(10,6))
plt.hist(math_scores, bins=5, color='lightgreen', edgecolor='black')
plt.title("Math Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Number of Students")
plt.show()
```

```python
# 6. Box Plot
plt.figure(figsize=(8,6))
plt.boxplot([math_scores, science_scores, english_scores], tick_labels=['Math', 'Science', 'English'])
plt.title("Score Spread by Subject")
plt.ylabel("Scores")
plt.show()
```

```python
# 7. Area Plot (Stackplot)
plt.figure(figsize=(12,6))
plt.stackplot(names, math_scores, science_scores, english_scores, labels=['Math','Science','English'], alpha=0.6)
plt.title("Cumulative Scores Area Plot")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Math Scores
plt.subplot(2,2,1)
plt.bar(names, math_scores, color='skyblue')
plt.title("Math Scores")

# Science Scores
plt.subplot(2,2,2)
plt.bar(names, science_scores, color='orange')
plt.title("Science Scores")

# English Scores
plt.subplot(2,2,3)
plt.bar(names, english_scores, color='green')
plt.title("English Scores")

# Average Scores
plt.subplot(2,2,4)
plt.plot(names, average_scores, marker='o', color='red')
plt.title("Average Scores")

plt.tight_layout()  # Automatically adjust spacing
plt.show()
```
